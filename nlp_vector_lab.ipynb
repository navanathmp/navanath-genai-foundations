```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv("Reviews.csv", nrows=50000)

print(df.shape)
```

```python
light_reviews = df[df["Text"].str.contains("light", case=False, na=False)]

print("Total reviews containing 'light':", len(light_reviews))
```

```python
for idx, review in light_reviews['Text'].head(50).items():
    print("\n")
    print(review)
    print("="*100)
```

```python
sentences = [
    "This is a confection that has been around a few centuries. It is a light, pillowy citrus gelatin with nuts.",
    
    "The packaging was broken and let too much light inside the box.",
    
    "It is not light and fluffy. It is very gooey."
]
```

```python
vectorizer = CountVectorizer()

X = vectorizer.fit_transform(sentences)

dense_matrix = X.toarray()

print(dense_matrix)
```

```python
vec_1 = dense_matrix[0]
vec_2 = dense_matrix[1]
vec_3 = dense_matrix[2]

print(vec_1)
print(vec_2)
print(vec_3)
```

```python
sim_12 = np.dot(vec_1, vec_2)

sim_13 = np.dot(vec_1, vec_3)

print("Similarity (Sentence 1 vs Sentence 2):", sim_12)

print("Similarity (Sentence 1 vs Sentence 3):", sim_13)
```
